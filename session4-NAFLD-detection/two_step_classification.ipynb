```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('../data/expression_data.csv', index_col=False)
data.head()
```

```python
data.set_index(data.columns[0])
```

```python
import pickle
with open('../Q2/selected_features.pkl', 'rb') as f:
    selected_features = pickle.load(f)
```

```python
selected_data = pd.concat([data[selected_features], data[['Simplified_class']]], axis=1)
```

```python
data.Simplified_class.unique()
```

```python
first_data = selected_data.copy()
func = lambda x: x if x not in ['Advanced_fibrosis', 'Non_advanced_Fibrosis'] else 'Abnormal'
first_data.Simplified_class = first_data.Simplified_class.apply(func)
```

```python
from sklearn.model_selection import train_test_split
X = first_data.drop('Simplified_class', axis=1)
y = first_data.Simplified_class
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=24)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(C=0.1, solver='lbfgs', max_iter=1000)
model.fit(X_train, y_train)
```

```python
from sklearn.metrics import classification_report
y_pred = model.predict(X_test)
print('First Classifier:')
print(classification_report(y_test, y_pred))
```

```python
second_data = selected_data.copy()
second_data = second_data[second_data['Simplified_class'] != 'Normal']
second_data.Simplified_class.value_counts()
```

```python
X2 = second_data.drop('Simplified_class', axis=1)
y2 = second_data.Simplified_class
X_train2, X_test2, y_train2, y_test2 = train_test_split(X2, y2, test_size=0.2, random_state=24)
model2 = LogisticRegression(C=0.1, solver='lbfgs', max_iter=1000)
model2.fit(X_train2, y_train2)
```

```python
y_pred2 = model2.predict(X_test2)
print('Second Classifier:')
print(classification_report(y_test2, y_pred2))
```

```python
import pickle
with open('first_model.pkl', 'wb') as f:
    pickle.dump(model, f)
with open('second_model.pkl', 'wb') as f:
    pickle.dump(model2, f)
```
