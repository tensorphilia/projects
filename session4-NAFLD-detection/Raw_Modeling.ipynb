```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('data/expression_data.csv')
data.head()
```

```python
data.set_index(data.columns[0], inplace=True)
data.head()
```

```python
# shuffling data to ensure uniform distribution and random sampling
data = data.sample(frac=1)
data.head()
```

```python
X = data.drop(columns=['Simplified_class'])
y = data['Simplified_class']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_temp, y_train, y_temp = train_test_split(X, y, random_state=42, test_size=0.2)
X_validation, X_test, y_validation, y_test = train_test_split(X_temp, y_temp, random_state=42, test_size=0.5)
```

```python
del X_temp, y_temp
```

# KNN

```python
from sklearn.neighbors import KNeighborsClassifier
knn_5 = KNeighborsClassifier(n_neighbors=5)
knn_5.fit(X_train, y_train)
```

```python
from sklearn.metrics import classification_report
y_pred_knn_5 = knn_5.predict(X_validation)
print(classification_report(y_true=y_validation, y_pred=y_pred_knn_5))
```

```python
# fine-tuning parameter
from sklearn.metrics import f1_score

best_f1 = 0
best_n_neighbors = 0
f1_scores = []
param_grid = range(1, 16)  

for n_neighbors in param_grid:
    knn = KNeighborsClassifier(n_neighbors=n_neighbors)
    knn.fit(X_train, y_train)
    y_validation_pred = knn.predict(X_validation)
    validation_f1 = f1_score(y_validation, y_validation_pred, average='weighted')
    f1_scores.append(validation_f1)
    if validation_f1 > best_f1:
        best_f1 = validation_f1
        best_n_neighbors = n_neighbors

print("Best n_neighbors found: ", best_n_neighbors)
print("Best validation F1 score: {:.2f}".format(best_f1))
```

```python
# plotting the result
plt.figure(figsize=(10, 6))
plt.plot(param_grid, f1_scores, marker='o')
plt.xlabel('Number of Neighbors (n_neighbors)')
plt.ylabel('Validation F1 Score')
plt.title('F1 Score vs. Number of Neighbors in KNN')
plt.grid()
plt.show()
```

```python
# final test on the test set
best_knn = KNeighborsClassifier(n_neighbors=best_n_neighbors)
best_knn.fit(X_train, y_train)
y_test_pred = best_knn.predict(X_test)
test_f1 = f1_score(y_test, y_test_pred, average='weighted')
print("Test set F1 score with best n_neighbors: {:.2f}".format(test_f1))
print(classification_report(y_test, y_test_pred))
```

# Logistic Regression

```python
from sklearn.linear_model import LogisticRegression

param_grid = {'C': np.logspace(-4, 4, 20)}
best_f1 = 0
best_C = 0
f1_scores = []

for C in param_grid['C']:
    log_reg = LogisticRegression(C=C, max_iter=1000)
    log_reg.fit(X_train, y_train)
    y_validation_pred = log_reg.predict(X_validation)
    validation_f1 = f1_score(y_validation, y_validation_pred, average='weighted')
    f1_scores.append(validation_f1)
    if validation_f1 > best_f1:
        best_f1 = validation_f1
        best_C = C

print("Best C found: ", best_C)
print("Best validation F1 score: {:.2f}".format(best_f1))
```

```python
# evaluation on the test set
best_log_reg = LogisticRegression(C=best_C, max_iter=1000)
best_log_reg.fit(X_train, y_train)
y_test_pred = best_log_reg.predict(X_test)
test_f1 = f1_score(y_test, y_test_pred, average='weighted')
print("Test set F1 score with best C: {:.2f}".format(test_f1))
```

```python
print(classification_report(y_test, y_test_pred))
```

# SVM

```python
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import f1_score, make_scorer
import seaborn as sns

# parameter grid for SVM
param_grid = {
    'C': np.logspace(-4, 4, 10),  # Regularization parameter
    'gamma': np.logspace(-4, 4, 10),  # Kernel coefficient for RBF kernel
    'kernel': ['rbf']  # Using RBF kernel
}
f1_scorer = make_scorer(f1_score, average='weighted')
grid_search = GridSearchCV(SVC(), param_grid, scoring=f1_scorer, cv=5, verbose=4)
grid_search.fit(X_train, y_train)
print("Best parameters found: ", grid_search.best_params_)
print("Best cross-validation F1 score: {:.2f}".format(grid_search.best_score_))
```

```python
# prediction on test set 
best_svm = grid_search.best_estimator_
y_test_pred = best_svm.predict(X_test)
test_f1 = f1_score(y_test, y_test_pred, average='weighted')
print("Test set F1 score: {:.2f}".format(test_f1))
print(classification_report(y_test, y_test_pred))
```

```python
# plot the results 
import seaborn as sns

# Extract results from grid search
results = grid_search.cv_results_
mean_test_scores = results['mean_test_score'].reshape(len(param_grid['C']), len(param_grid['gamma']))

# Plot heatmap of mean test scores
plt.figure(figsize=(10, 6))
sns.heatmap(mean_test_scores, xticklabels=param_grid['gamma'], yticklabels=param_grid['C'], cmap='viridis')
plt.xlabel('gamma')
plt.ylabel('C')
plt.title('Validation F1 Score for each combination of C and gamma')
plt.show() 
print(f'Best Parameters: {grid_search.best_params_}')
```

# Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB
param_grid = {'var_smoothing': np.logspace(-9, -1, 10)} 
f1_scorer = make_scorer(f1_score, average='weighted')

# Create the GridSearchCV object
grid_search = GridSearchCV(GaussianNB(), param_grid, scoring=f1_scorer, cv=5, verbose=2)

# Fit the model on the training set
grid_search.fit(X_train, y_train)

# Print best parameters and score
print("Best parameters found: ", grid_search.best_params_)
print("Best cross-validation F1 score: {:.2f}".format(grid_search.best_score_))
```

```python
# test set evaluation
best_nb = grid_search.best_estimator_
y_test_pred = best_nb.predict(X_test)
test_f1 = f1_score(y_test, y_test_pred, average='weighted')
print("Test set F1 score: {:.2f}".format(test_f1))
print(classification_report(y_test, y_test_pred))
```

# Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
param_grid = {
    'max_depth': [3, 5, 7, 9, None],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'max_features': [None, 'sqrt', 'log2']
}
f1_scorer = make_scorer(f1_score, average='weighted')
grid_search = GridSearchCV(DecisionTreeClassifier(), param_grid, scoring=f1_scorer, cv=5, verbose=4)
grid_search.fit(X_train, y_train)
print("Best parameters found: ", grid_search.best_params_)
print("Best cross-validation F1 score: {:.2f}".format(grid_search.best_score_))
```

```python
best_dt = grid_search.best_estimator_
y_test_pred = best_dt.predict(X_test)
test_f1 = f1_score(y_test, y_test_pred, average='weighted')
print("Test set F1 score: {:.2f}".format(test_f1))
print(classification_report(y_test, y_test_pred))
```

# Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
param_grid = {
    'n_estimators': [50, 100, 200],
    'max_depth': [5, 10, 20],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'max_features': ['sqrt', 'log2', None]
}
f1_scorer = make_scorer(f1_score, average='weighted')
grid_search = GridSearchCV(RandomForestClassifier(), param_grid, scoring=f1_scorer, cv=5, verbose=4)
grid_search.fit(X_train, y_train)
print("Best parameters found: ", grid_search.best_params_)
print("Best cross-validation F1 score: {:.2f}".format(grid_search.best_score_))
```

```python
params = dict(max_depth=5, max_features=None, min_samples_leaf=2, min_samples_split=2, n_estimators=50)
random_forest = RandomForestClassifier(**params)
random_forest.fit(X_train, y_train)
y_test_pred = random_forest.predict(X_test)
print(classification_report(y_test, y_test_pred))
```

```python
# searching again
param_grid = {
    'n_estimators': [50, 100, 200],
    'max_depth': [5, 10, 20],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'max_features': ['sqrt', 'log2']
}
f1_scorer = make_scorer(f1_score, average='weighted')
grid_search = GridSearchCV(RandomForestClassifier(), param_grid, scoring=f1_scorer, cv=5, verbose=4)
grid_search.fit(X_train, y_train)
print("Best parameters found: ", grid_search.best_params_)
print("Best cross-validation F1 score: {:.2f}".format(grid_search.best_score_))
```

```python
best_rf = grid_search.best_estimator_
y_test_pred = best_rf.predict(X_test)
test_f1 = f1_score(y_test, y_test_pred, average='weighted')
print("Validation set F1 score: {:.2f}".format(test_f1))
print(classification_report(y_test, y_test_pred))
```

```python
# Extract results from grid search
results = grid_search.cv_results_

# Plotting n_estimators vs. mean_test_score
plt.figure(figsize=(10, 6))
sns.boxplot(x=results['param_n_estimators'].data.astype(str), y=results['mean_test_score'])
plt.xlabel('n_estimators')
plt.ylabel('Mean Cross-Validation F1 Score')
plt.title('F1 Score vs. n_estimators in RandomForestClassifier')
plt.show()
```
